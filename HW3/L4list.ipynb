# Stack using Python List
# Stack has unlimited size capacity

```python
############################################################
# Stack.py 
# Implements Stack object
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2022
###########################################################

###########################################################
#  class  Stack
# Our stack is infinite size
###########################################################   
class Stack():
    def __init__(self): 
        #MUST USE ONLY PYTHON LIST
        self._a = []
        #print("WRITE CODE")
        self._maxspace = 0
        self._currentspace = 0
        
    def __len__(self):
        return(len(self._a))
        
    def push(self,T)->"None":
        self._currentspace += 1
        if(self._maxspace < self._currentspace):
            self._maxspace = self._currentspace
        self._a.append(T)
    
    def pop(self)->"int":
        if(self.empty()):
            return None
        else:
            self._currentspace -= 1
            return(self._a.pop())
    
    def top(self)->"int":
        if(self.empty()):
            return(None)
        else:
            return(self._a[-1])
    
    def empty(self)->"bool":
        if(len(self._a)==0):
            return True
        else:
            return False
    
    def space(self):
        #print(self._maxspace)
        return(self._maxspace)

 
```

# Circular queue using Python list 
# Circular queue has limited size capacity

```python
############################################################
# Queue.py 
# Implements Queue object
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2022
###########################################################

###########################################################
#  class Queue
# Our Queue is finite size k
###########################################################   
class Queue():
    def __init__(self,k:'int'):
        # You must use Python List
        # Must use all spaces
        self._a = [None]*k #CANNOT CHANGE THIS. k space is already allocated
        self._MAX = k
        ## YOU CAN HAVE YOUR PRIVATE DATA MEMBER HERE
        self._front = -1
        self._back = -1

        
    def enQueue(self, T)->'bool':
        ## YOU CANNOT CALL append in this routine as U already have enough space
        ## YOU CAN DO: self._a[pos] = T #pos is some position between 0 to self._MAX-1
        #print("WRITE CODE")
        queue_length = len(self)
        self._back += 1
        if(queue_length==self._MAX):
            self._back -= 1
            return(False)
        if(self._back == self._MAX):
            self._back = 0
        if(self._front==-1):
            self._front = 0
        self._a[self._back] = T
        return(True)
        
    def deQueue(self)->"bool":
        if(self._front == -1):
            return(False)
        if(self._front==self._back):
            self._front = -1
            self._back = -1
        else:
            self._front += 1
            if(self._front == self._MAX):
                self._front = 0
        return(True)
           
        
    def Front(self)->'T':
        ## YOU CANNOT CALL pop(0). NOTE: pop(0) is O(n). We want THETA(1)
        #print("WRITE CODE")
        if(self._front == -1):
            return(-1)
        return(self._a[self._front])
    
    def Rear(self)->"int":
        if(self._back < 0):
            return(-1)
        #print(self._back)
        return(self._a[self._back])

        
    ## WRITE ALL OTHER ROUTINES
    def __len__(self)->'int':
        #print("WRITE CODE to return number of elements in Queue at this point")
        if(self._front == -1):
            return(0)
        count = 0
        if(self._back < self._front):
            for i in range(0,self._back+1):
                count += 1
            for i in range(self._front,self._MAX):
                count += 1
            return(count)
        else:
            return(self._back-self._front+1)
        
    #Print from FRONT TO REAR
    def __str__(self)->'string':
        s = ""
        ## WRITE CODE HERE
        if(len(self)==0):
            return(s)
        if(self._back < self._front):
            for i in range(0,self._back+1):
                s+= str(self._a[i])+" "
            for i in range(self._front,self._MAX):
                s+= str(self._a[i])+" "
        else:
            for i in range(self._front,self._back+1):
                s = s +str(self._a[i])+" "
        return(s)

    
    def isEmpty(self):
        if(self._front == -1 and self._back == -1):
            return(True)
        else:
            return(False)

    def isFull(self)->'bool':
        if(len(self)==self._MAX):
            return(True)
        else:
            return(False) 
        
        
    
```

```python
############################################################
# StackUsingQueue.py 
# Implements Stack object using Queue
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2022
###########################################################

############################################################
#  All imports here
###########################################################
#import sys # For getting Python Version
#from Queue import *

###########################################################
#  class  Stack using Queue
#Implement the MyStack class:

#MUST USE ONLY THE FOLLOWING FUNCTIONS FROM QUEUE

#You must use only standard operations of a queue, 
#which means that only push to back, peek/pop from front, size and is empty operations are valid.

#1. enQueue(self, T)->'bool'
#2. deQueue(self)->'bool'
#3. Front(self)->'T'
#4  Rear(self)->'T':
#5. isEmpty(self)->'Bool':
#6. len(q) - numner of elements int

#Our StackUsingQueue is finite size n
###########################################################   
class StackUsingQueue():
    def __init__(self, n:'size'=10): 
        # ONLY DATA STRUCTURE YOU CAN USE HERE IS ONLY QUEUE THAY YOU WROTE
        #print("WRITE CODE")
        self._queue = Queue(n)
    def __len__(self):
        return(len(self._queue))
    
    def push(self,T)->'None':
        if(self._queue.isFull()):
            return(None)
        if(self._queue.isEmpty()):
            self._queue.enQueue(T)
            return()
        self._queue.enQueue(T)
        for i in range(len(self)-1):
            front = self._queue.Front()
            self._queue.deQueue()
            self._queue.enQueue(front)

    def pop(self)->'T':
        front = self._queue.Front()
        self._queue.deQueue()
        return(front)
    
    def top(self)->'int':
        if(len(self)==0):
            return(None)
        else:
            return(self._queue.Front())
        
    def empty(self)->'bool':
        return(self._queue.isEmpty())
    
    def is_full(self)->'bool':
        return(self._queue.isFull())
    


        
```

```python
############################################################
# QueueUsingStack.py 
# Implements Queue object using Stack
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2022
###########################################################

############################################################
#  All imports here
###########################################################
import sys # For getting Python Version
##Comment in Notebook
#from Stack import *

###########################################################
#  class  Queue using Stack

# push(x) void push(T x) Pushes element T to the back of the queue.
# T = pop() T pop() Removes the element from the front of the queue and returns it.
# T = peek() T peek() Returns the element at the front of the queue.
# bool empty()  Returns true if the queue is empty, false otherwise.
###########################################################

class QueueUsingStack():
    def __init__(self): 
        # ONLY DATA STRUCTURE YOU CAN USE HERE IS ONLY STACK THAT YOU WROTE
        #print("WRITE CODE")
        self._stack1 = Stack()
        self._stack2 = Stack()
        self._currentspace = 0
        self._maxspace = 0

    def __len__(self):
        return(len(self._stack1)+len(self._stack2))   

    def push(self,T)->"None":
        self._stack1.push(T)
        self._currentspace += 1
        if(self._maxspace < self._currentspace):
            self._maxspace = self._currentspace

    def pop(self)->"int":
        if(self._stack1.empty() and self._stack2.empty()):
            return(None)
        if((self._stack1.empty()==False) and self._stack2.empty()):
            while(self._stack1.empty()==False):
                self._stack2.push(self._stack1.pop())
        self._currentspace -= 1
        return(self._stack2.pop())
    
    def top(self)->"int":
        if(self._stack1.empty() and self._stack2.empty()):
            return(None)
        if((self._stack1.empty()==False) and self._stack2.empty()):
            while(self._stack1.empty()==False):
                self._stack2.push(self._stack1.pop())
        return(self._stack2.top())
        
    def peek(self):
        return(self.top())

    def space(self):
        return(self._maxspace)
    
    def empty(self):
        if(len(self)==0):
            return(True)
        else:
            return(False)


            
```

```python
############################################################
# MinStack.py 
# Implements MinStack
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2022
###########################################################
############################################################
# MinStack.py 
# Implements MinStack
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2022
###########################################################

############################################################
#  All imports here
###########################################################
import sys # For getting Python Version
##Comment in Notebook
#from Stack import *

###########################################################
#  Min Stack

# push(x) void push(T x) Pushes element T to the MinStack
# void pop() void pop() Removes the element from the MinStack
# T = top() T pop() Returns the front element 
# T = getMin() Returns MIN T in THETA(1) time
###########################################################
class MinStack():
    def __init__(self): 
        # ONLY DATA STRUCTURE YOU CAN USE HERE IS ONLY STACK THAT YOU WROTE
        #print("WRITE CODE")
        self._stack = Stack()
        self._min_element = None
    
    def getMin(self):
        if(self._stack.empty()):
            return None
        else:
            return(self._min_element)
                
    def push(self,T)->"None":
        if(self._stack.empty()):
            self._stack.push(T)
            self._min_element = T
        else:
            if(T >= self._min_element):
                self._stack.push(T)
            else:
                self._stack.push((2*T)-self._min_element)
                self._min_element = T

    def top(self)->"int":
        if(self._stack.empty()):
            return(None)
        if(self._stack.top()>self._min_element):
            return(self._stack.top())
        else:
            return(self._min_element)

    def pop(self)->"int":
        if(self._stack.empty()):
            return(None)
        if(self._stack.top()>self._min_element):
            return(self._stack.pop())
        else:
            current_min = self._min_element
            self._min_element = 2*(self._min_element) - self._stack.top()
            self._stack.pop()
            return(current_min)
            
    def empty(self):
        return(self._stack.empty())


```
